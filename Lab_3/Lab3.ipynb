```python
import numpy as np
from scipy.integrate import odeint, solve_ivp
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

Рівняння Ван дер Поля

```python
w = 1
k = 1
mu = 0.8
la = 1
def VanDerPole(y, t):
    x, v = y
    return [v, -w**2*x + mu*(1-x**2)*v + k*x*np.cos(la*t)]
t = np.linspace(0,40*np.pi,1000)
result = odeint(VanDerPole, [0,1], t)
x = result[:,0]
v = result[:,1]
```

```python
plt.plot(t,x,lw=2)
plt.xlabel('t')
plt.ylabel('x')
plt.xlim(0,40*np.pi)
plt.grid()
```

```python
plt.plot(x,v,lw=2)
plt.xlabel('x')
plt.ylabel('v')
plt.xlim(-3,3)
plt.ylim(-5,5)
plt.grid()
```

Рівняння Матье

```python
w = 0.3
e = 0.1
lamb = 1
def mathew(y, t):
    x, v = y
    return [v, -(w**2 + e*np.cos(lamb*t))*x]
t = np.linspace(0,40*np.pi,1000)
result = odeint(mathew, [0,1], t)
x = result[:,0]
v = result[:,1]
```

```python
plt.plot(t,x,lw=2)
plt.xlabel('t')
plt.ylabel('x')
plt.xlim(0,40*np.pi)
plt.grid()
```

```python
plt.plot(x,v,lw=2)
plt.xlabel('x')
plt.ylabel('v')
plt.xlim(-5,5)
plt.ylim(-2.5,2.5)
plt.grid()
```

Рівняння Дюфінга

```python
w = 1
k = 1
e = 0
la = 1
def duffing(y, t):
    x, v = y
    return [v, -w**2*x - e*x**3 + k*np.cos(la*t)]
t = np.linspace(0,40*np.pi,1000)
result = odeint(duffing, [0,1], t)
x = result[:,0]
v = result[:,1]
```

```python
plt.plot(t,x,lw=2)
plt.xlabel('t')
plt.ylabel('x')
plt.xlim(0,40*np.pi)
plt.grid()
```

```python
plt.plot(x,v,lw=2)
plt.xlabel('x')
plt.ylabel('v')
plt.xlim(-75,75)
plt.ylim(-75,75)
plt.grid()
```

# Маятник

```python
import numpy as np
from scipy.integrate import odeint, solve_ivp
import matplotlib.pyplot as plt
from scipy.integrate import odeint
m = 1
k = 30
l = 2
g = 9.81
def f(initial, t):
    r, R, phi, Phi = initial           
    return [R, r*Phi**2 + g*np.cos(phi)-k*(r-l)/m, Phi, (-g*np.sin(phi) - 2*R*Phi)/r]

t = np.arange(0, 20, 0.001)

res = odeint(f, [2, 0, np.pi/2, 0], t)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, polar=True)
ax.plot(res[:,2], res[:,0])
ax.set_theta_zero_location("S")
```
